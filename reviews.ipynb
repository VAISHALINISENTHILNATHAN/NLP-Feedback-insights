```python
import random
import pandas as pd

# A set of templates and phrases to generate synthetic reviews
positive_phrases = [
    "I love this product",
    "Absolutely fantastic",
    "This exceeded my expectations",
    "Very happy with my purchase",
    "The quality is amazing",
    "Great value for money",
    "Highly recommend this",
    "Will definitely buy again",
    "Customer support was helpful",
    "Really well made"
]

negative_phrases = [
    "Terrible experience",
    "I am disappointed",
    "Broke after",
    "Very poor quality",
    "Not worth the price",
    "Customer support did not help",
    "Delivery was late",
    "Battery drains too fast",
    "Wrong item delivered",
    "Not what I expected"
]

neutral_phrases = [
    "Itâ€™s okay",
    "Nothing special",
    "Does the job",
    "Average product",
    "Meets my needs",
    "Could be better",
    "Fair quality",
    "Decent performance",
    "Not bad, not great",
    "Just fine"
]

def make_review():
    sentiment = random.choices(
        ["positive", "negative", "neutral"],
        weights=[0.5, 0.3, 0.2],
        k=1
    )[0]
    
    if sentiment == "positive":
        phrase = random.choice(positive_phrases)
        extra = random.choice([
            "and works perfectly.", 
            "and I am really satisfied.", 
            "and the design is very stylish.", 
            "with no issues so far."
        ])
    elif sentiment == "negative":
        phrase = random.choice(negative_phrases)
        extra = random.choice([
            "after only a week.",
            "and I'm very upset.",
            "so I will not buy again.",
            "which is totally unacceptable."
        ])
    else:
        phrase = random.choice(neutral_phrases)
        extra = random.choice([
            "but nothing extraordinary.",
            "for the price.",
            "as I expected.",
            "and it's okay."
        ])
    
    review = f"{phrase} {extra}"
    return review

# Generate 100 reviews
reviews = [make_review() for _ in range(100)]

# Put into a DataFrame
df = pd.DataFrame({"review": reviews})

# Save to CSV
import os
os.makedirs("data/raw", exist_ok=True)
df.to_csv("data/raw/reviews.csv", index=False)

print("Generated synthetic reviews saved to data/raw/reviews.csv")
df.head(10)
```
